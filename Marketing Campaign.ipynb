# Hypotheses Testing

```python
data.dtypes
```

```python
X = data[['AcceptedCmp1','AcceptedCmp2','AcceptedCmp3','AcceptedCmp4','AcceptedCmp5']]
```

```python
y = data['Response']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.30,random_state=50)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
X_test.shape
```

```python
y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
linreg = LinearRegression()
```

```python
linreg.fit(X_train,y_train)
```

```python
print(linreg.intercept_)
```

```python
coeff = pd.DataFrame(linreg.coef_,X.columns,columns=['Coefficient'])
```

```python
coeff
```

```python
from sklearn.feature_selection import f_regression
```

```python
fregression = f_regression(X_train,y_train)
```

```python
fregression
```

# Since the P-value is less than 0.05, we reject the null hypothesis and accept the alternative hypothesis

```python
y_pred = linreg.predict(X_test)
```

```python
y_pred
```

```python
y_test
```

```python
from sklearn import metrics
```

```python
print('RSME:',np.sqrt(metrics.mean_squared_error(y_test,y_pred)))
```

```python
print('R Squared:',metrics.r2_score(y_test,y_pred))
```

### The R squared value is close to 0, thus the data points doesn't fit properly
